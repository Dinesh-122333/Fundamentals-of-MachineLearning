```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic graph

```python
#assign elements to variable
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.show()#where it only show the graph
```

```python
#Adding lable to graph
x=[1,3,4]
y=[4,5,6]
plt.title("First Graph") 
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()  

```

```python
#adding some attribute to title
#also we can change the size of the label  
x=[1,3,4]
y=[4,5,6]
plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 
plt.plot(x,y)
plt.xlabel("X Axis",fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.ylabel("Y Axis",fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.show() 
```

```python
#shows in integer value as we entered in x
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y)

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,3,4])
plt.yticks([0,4,5,6,11])
plt.show()  

```

```python
# labeling the line segment
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y ,label='2x')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,3,4])
plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()  

```

```python
#colour of the line
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y ,label='2x',color='green')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,3,4])
plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()  

```

```python
#adding linewidth and dots,symbol,etc,.. and size of the dots, color of the dots 
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y ,label='2x',color='green',linewidth='2',marker='.',markersize='15',markeredgecolor='yellow')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,3,4])
plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()
```

```python
#style of the line 
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y ,label='2x',color='green',linewidth='2',linestyle='--',marker='.',markersize='15',markeredgecolor='yellow')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,3,4])
plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()
```

```python
#short hand notation
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y,'g*:' ,label='2x')#fmt='[color][marker][line]'

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,3,4])
plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()
```

##Two line segment


```python
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y ,label='2x',color='green',linewidth='2',marker='.',markersize='15',markeredgecolor='yellow')


#second line
x2=np.arange(0,5,2)

plt.plot(x2,x2**2,'k*:',label='X^2')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
#plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()
```

```python
#making two different format of line for one line
x=[0,1,2,3,4]
y=[0,4,5,6,11]
plt.plot(x,y ,label='2x',color='green',linewidth='2',marker='.',markersize='15',markeredgecolor='yellow')


#second line
x2=np.arange(0,5,0.5)

plt.plot(x2[:5],x2[:5]**2,'k',label='X^2')
plt.plot(x2[4:],x2[4:]**2,'r--')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
#plt.yticks([0,4,5,6,11])

plt.legend()
plt.show()
```

```python
#resize of the graph
x=[0,1,2,3,4]
y=[0,4,5,6,11]


plt.figure(figsize=(6,4),dpi=150)#dpi=pixel per inch ,which is used to show the clarity of graph
plt.plot(x,y ,label='2x',color='green',linewidth='2',marker='.',markersize='15',markeredgecolor='yellow')


#second line
x2=np.arange(0,5,0.5)

plt.plot(x2[:5],x2[:5]**2,'k',label='X^2')
plt.plot(x2[4:],x2[4:]**2,'r--')

plt.title("First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20}) 

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xticks([0,1,2,3,4])
#plt.yticks([0,4,5,6,11])
#saving the graph
plt.savefig('mygraph1.png',dpi=200)

plt.legend()
plt.show()
```

## Barchart


```python
label=['A','B','C']
values=[1,3,5]
plt.bar(label,values)
plt.show()
```

```python
#styles in bar
label=['A','B','C']
values=[1,3,5]

bars=plt.bar(label,values)
bars[0].set_hatch('/')#the slash only for a first bar
bars[1].set_hatch('o')#the slash only for a second bar
bars[2].set_hatch('*')#the slash only for a third bar

plt.figure(figsize=(6,5))
plt.show()
```

```python
#short hand for pattern
label=['A','B','C']
values=[1,3,5]

bars=plt.bar(label,values)
pattern=['/','o','*']
for bar in bars:
    bar.set_hatch(pattern.pop(0))


plt.show()
```

```python
#from last to first
label=['A','B','C']
values=[1,3,5]

bars=plt.bar(label,values)
pattern=['/','o','*']
for bar in bars:
    bar.set_hatch(pattern.pop())


plt.show()
```
