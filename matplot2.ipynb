```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#combination of pandas and matplotlib

gas=pd.read_csv('gas_prices.csv')

plt.figure(figsize=(7,5),dpi=150)

plt.title('Gas Price')
plt.plot(gas.Year,gas.Australia,'b*--',label='US')
plt.plot(gas.Year,gas.Italy,'k^:',label='Italy')
plt.plot(gas.Year, gas['South Korea'],'r.-',label='South Korea')


plt.xlabel('Year')
plt.ylabel('$Dollar$')
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
#for multiple data
gas=pd.read_csv('gas_prices.csv')

plt.figure(figsize=(7,5),dpi=150)

plt.title('Gas Price')
for country in gas:#printing all data in gas_csv file
    if country != 'Year':#to print the country's only
        plt.plot(gas.Year,gas[country],marker='.')
        

plt.xlabel('$Dollar$')
plt.ylabel('Year')
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
#only particular country 
gas=pd.read_csv('gas_prices.csv')

plt.figure(figsize=(7,5),dpi=150)

plt.title('Gas Price')
country_to_see=['USA','Italy','South Korea','Canada','Germany']
for country in country_to_see:#printing all data in gas_csv file
    if country != 'Year':#to print the country's only
        plt.plot(gas.Year,gas[country],marker='.')
        

plt.xlabel('year')
plt.ylabel('$dollar$')
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head(5)
```

## Histogram


```python
bins=[40,50,60,70,80,90,100]

plt.title('Fifa player skill 2018')
plt.xlabel('Skill of player')
plt.ylabel('Number of player')

plt.hist(fifa.Overall,bins=bins,color='#afffff')#change the colour also
plt.xticks(bins)


plt.show()
```

## Pie chart


```python
#pecentage of player is in left or right foot
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
plt.pie([left,right])
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
label=['left','right']
color=['#abbfdc','#abcdef']
#for percentage sign we need two percentage symbol
plt.pie([left,right],labels=label,colors=color,autopct='%.2f %%')#autopct mean auto percentege where it print only the starting two decimal value 
plt.title('Football player foot preference')
plt.show()
```

```python
#another pie chart
#we can take weight data
#in that case we need to print without lbs
#we need to convert to integer also
fifa.Weight=[int(x.strip('lbs'))if type(x)==str else x for x in fifa.Weight]#because it also has float value to so we conditioning them as using if statement


light=fifa.loc[(fifa.Weight<125)].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight< 150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
light_hev=fifa.loc[(fifa.Weight>=175) & (fifa.Weight>200)].count()[0]
heavy=fifa.loc[(fifa.Weight>=200)].count()[0]




weights=[light,light_medium,medium,light_hev,heavy]
labels=['Light','Light Medium','Medium','Light Heavy','Heavy']
color=['Red','#aaabca','#aabbcc','#aaaabf','#ababab']


plt.pie(weights,labels=labels,colors=color)
plt.show()
```

```python
#by default style of colour 
fifa.Weight=[int(x.strip('lbs'))if type(x)==str else x for x in fifa.Weight]#because it also has float value to so we conditioning them as using if statement
plt.style.use('ggplot')

light=fifa.loc[(fifa.Weight<125)].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight< 150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
light_hev=fifa.loc[(fifa.Weight>=175) & (fifa.Weight>200)].count()[0]
heavy=fifa.loc[(fifa.Weight>=200)].count()[0]




weights=[light,light_medium,medium,light_hev,heavy]
labels=['Light','Light Medium','Medium','Light Heavy','Heavy']
color=['Red','#aaabca','#aabbcc','#aaaabf','#ababab']


plt.pie(weights,labels=labels,autopct='%.2f %%')
plt.show()
```

```python
#for avoiding coliding between percentage
fifa.Weight=[int(x.strip('lbs'))if type(x)==str else x for x in fifa.Weight]#because it also has float value to so we conditioning them as using if statement
plt.style.use('ggplot')

light=fifa.loc[(fifa.Weight<125)].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight< 150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
light_hev=fifa.loc[(fifa.Weight>=175) & (fifa.Weight>200)].count()[0]
heavy=fifa.loc[(fifa.Weight>=200)].count()[0]




weights=[light,light_medium,medium,light_hev,heavy]
labels=['Light','Light Medium','Medium','Light Heavy','Heavy']
color=['Red','#aaabca','#aabbcc','#aaaabf','#ababab']
#its used to split the piechart into pieces
explode=(.4,.2,.0,.3,.4)

plt.title('Weight of the player in FIFA 2018',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.pie(weights, labels=labels , autopct='%.2f %%', pctdistance=0.8 ,explode=explode)
plt.show()
```

## Box and whisker plot


```python
barcelona=fifa[fifa.Club == 'FC Barcelona']['Overall']
madrid=fifa[fifa.Club == 'Real Madrid']['Overall']

plt.style.use('default')

labels=['FC barcelona','Real Madrid']

#patch_artist is for making the inside colour of the box 
#medianprops for median line width
boxes=plt.boxplot([barcelona,madrid], labels=labels,patch_artist=True ,medianprops={'linewidth':2})
for box in boxes['boxes']:
    #outline of the box colour
    box.set(color='#4dffff' ,linewidth=2)
    #inside of the box colour
    box.set(facecolor='#ccffff')




plt.title('Foot ball team comparision',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.ylabel('FIFA Overall Rating')
plt.show()
```
